Function 2) Define a function called vocab_size(text) that has a single parameter for the text, and which returns the vocabulary size of the text.

```python
def vocab_size(text):   #keyword def function(input parameter)
    return len(set(word.lower() for word in text))

#making all words lower case so that upper and lower case versions 
#of the same word aren't counted twice 

vocab_size(text7)
```

Function 4) Define a function find_language() that takes a string as its argument, and returns a list of languages that have that string as a word. Use the udhr corpus and limit your searches to files in the Latin-1 encoding.

```python
from nltk.corpus import udhr


def find_language(w):
    languages = []
    for fileid in udhr.fileids():
        if w.lower() in list(map(str.lower, udhr.words(fileid))) and fileid.endswith('-Latin1'):
            languages.append(fileid[:-7])
    return languages        

#find_language('st')
#find_language('can')
#find_language('when')
#find_language('JULISTUS')
find_language('julistus')
```

Function 5) Define a function supergloss(s) that takes a synset s as its argument and returns a string consisting of the concatenation of the definition of s, and the definitions of all the hypernyms and hyponyms of s.

Synsets are set of synonyms that have the same meaning
Hypernyms are words whose meaning includes a group of other words, so bird and pet would be hypernyms of parrot 
A hyponym is a word whose meaning is included in the meaning of another word so dog, cat, panther are all hyponyms of "animal"

nltk includes WordNet for English, which has certain functions for definitions, synsets, hypernyms and hyponyms 

```python
from nltk.corpus import wordnet 

def supergloss(s):
    worddef = "definition: " + s.definition() + "\n\n"
    worddef = worddef + "Hypernyms:\n"
    for hypernym in s.hypernyms():
        worddef = worddef + hypernym.name() + ": " + hypernym.definition() + "\n"
    worddef = worddef + "\n" +"Hyponyms:\n"
    for hyponym in s.hyponyms():
        worddef = worddef + hyponym.name() + ": " + hyponym.definition() + "\n"
    return worddef

#print(supergloss(wordnet.synset('beagle.n.01')))
print(supergloss(wordnet.synset('dog.n.01')))
    
```

NLTK has multiple similarity measures, including path similarity which returns a score to denote how similar the senses of two words are. This similarity is measured based on the shortest path between the two words in the taxonomy net that includes hypernyms and hyponyms. This will give a score between 0 and 1. Other such similarity measures are the Leacock-Chodorow, Wu-Palmer, Resnik, Jiang-Conrath and Lin similarities. Each return a score to denote how similar two word senses but are based on different grounds such as information content along with other elements. 

This function has been created to analyse the score rankings of a set of word pairs given by Miller and Charles, 1998, in decreasing order. The different types of similarity measures can be used as they are available in NLTK.

```python
from operator import itemgetter   #to help in sorting the list in ascending order according to the third item in the list

word_pairs = [('car', 'automobile'), ('gem', 'jewel'), ('journey', 'voyage'), ('boy', 'lad'), ('coast', 'shore'), 
         ('asylum', 'madhouse'), ('magician', 'wizard'), ('midday', 'noon'), ('furnace', 'stove'), ('food', 'fruit'), 
         ('bird', 'cock'), ('bird', 'crane'), ('tool', 'implement'), ('brother', 'monk'), ('lad', 'brother'), 
         ('crane', 'implement'), ('journey', 'car'), ('monk', 'oracle'), ('cemetery', 'woodland'), ('food', 'rooster'), 
         ('coast', 'hill'), ('forest', 'graveyard'), ('shore', 'woodland'), ('monk', 'slave'), ('coast', 'forest'), 
         ('lad', 'wizard'), ('chord', 'smile'), ('glass', 'magician'), ('rooster', 'voyage'), ('noon', 'string')]
empli = []

#analysing the path similarity, can use lch, res, wup, jcn, lin similarities too
for word1, word2 in word_pairs:
    empli.append((word1, word2, wordnet.path_similarity(wordnet.synsets(word1)[0], wordnet.synsets(word2)[0])))
    
#sorted(empli, key=itemgetter(2))
sorted(empli, key=itemgetter(2), reverse=True)
```

Function 7) Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.

```python
import nltk
from nltk.corpus import *
gutenberg

whitman = nltk.Text(nltk.corpus.gutenberg.words('whitman-leaves.txt'))

def fifty_bigrams(text):
    stopwords = nltk.corpus.stopwords.words('english')
    bigrams = [b for b in nltk.bigrams(text) if b[0] not in stopwords and b[1] not in stopwords and any(w.isalpha() for w in b[0]) and any(w.isalpha() for w in b[1])]
    freqdist = nltk.FreqDist(bigrams)
    return [b for b in freqdist.most_common(50)]

fifty_bigrams(whitman)
```
